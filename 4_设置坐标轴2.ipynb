```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
x = np.linspace(-3,3,100)
y1 = 2*x +1
y2 = x**2

new_tick = np.linspace(-2,2,11)
print(new_tick)

#设置显示的范围

plt.xlim((-1,2)) #x在-1到2范围内
plt.ylim((-2,3)) #y在-2到3范围内

#对坐标轴描述
plt.xlabel('I AM X')
plt.ylabel('I AM Y')

plt.plot(x,y1,color = "red", linewidth = 1.0, linestyle = "--")
plt.plot(x,y2,color = "blue", linewidth = 5.0, linestyle = "-")

plt.xticks(new_tick)
plt.yticks([-1,0,1,2,3],['level1','level2','level3','level4','level5'])

#gca get current axis
ax = plt.gca()

#把右边和上边框去掉
ax.spines['top'].set_color('none')  # 设置上‘脊梁’为红色
ax.spines['right'].set_color('none')  # 设置上‘脊梁’为红色

#设置x刻度为'bottom'
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#设置坐标轴对应到0 1点
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',1))

plt.show()


```
