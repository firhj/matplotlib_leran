```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
x = np.linspace(-3,3,100)
y1 = 2*x +1
y2 = x**2

new_tick = np.linspace(-2,2,11)
print(new_tick)

#设置显示的范围

plt.xlim((-1,2)) #x在-1到2范围内
plt.ylim((-2,3)) #y在-2到3范围内

#对坐标轴描述
plt.xlabel('I AM X')
plt.ylabel('I AM Y')

l1, = plt.plot(x,y1,color = "red", linewidth = 1.0, linestyle = "--")
l2, = plt.plot(x,y2,color = "blue", linewidth = 5.0, linestyle = "-")
plt.legend( handles = [l1, l2]
           ,labels = ['test1','test2']
           ,loc = 'best')


plt.xticks(new_tick)
plt.yticks([-1,0,1,2,3],['level1','level2','level3','level4','level5'])

#gca get current axis
ax = plt.gca()

#把右边和上边框去掉
ax.spines['top'].set_color('none')  # 设置上‘脊梁’为红色
ax.spines['right'].set_color('none')  # 设置上‘脊梁’为红色


plt.show()
```
