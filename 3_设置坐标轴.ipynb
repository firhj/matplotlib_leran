```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
x = np.linspace(-3,3,100)
y1 = 2*x +1
y2 = x**2

#设置显示的范围

plt.xlim((-1,2)) #x在-1到2范围内
plt.ylim((-2,3)) #y在-2到3范围内

#对坐标轴描述
plt.xlabel('I AM X')
plt.ylabel('I AM Y')

plt.plot(x,y1,color = "red", linewidth = 1.0, linestyle = "--")
plt.plot(x,y2,color = "blue", linewidth = 5.0, linestyle = "-")
plt.show()
```
