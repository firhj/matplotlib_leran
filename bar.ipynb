```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
x = np.arange(10)
y = 2**x+10
plt.bar(x,y)
plt.show()
```

```python
x = np.arange(10)
y = 2**x+10
plt.bar(x,-y)
plt.show()
```

```python
x = np.arange(10)
y = 2**x+10
plt.bar(x,y,facecolor = 'red',edgecolor ='white')
plt.show()
```

```python
x = np.arange(10)
y = 2**x+10
plt.bar(x,y,facecolor = 'red',edgecolor ='white')
for x,y in zip(x,y):
    plt.text(x+0.1,y,'%.2f'% y, ha = 'center', va = 'bottom')
plt.show()
```

```python
x = np.arange(10)
y = 2**x+10
plt.bar(x,y,facecolor = 'red',edgecolor ='white')
for x,y in zip(x,y):
    plt.text(x+0.1,y,'%.2f'% y, ha = 'center', va = 'top')
plt.show()
```
