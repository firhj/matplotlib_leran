```python

import matplotlib.pyplot as plt

import numpy as np
```

```python
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
    
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)

X,Y = np.meshgrid(x,y)
plt.contourf(X,Y,f(X,Y)
             ,8 # 8+1条等高线
             ,alpha = 0.75
             ,cmap = plt.cm.hot #颜色
             )

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
    
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)

X,Y = np.meshgrid(x,y)
plt.contourf(X,Y,f(X,Y)
             ,8 # 8+1条等高线
             ,alpha = 0.75
             ,cmap = plt.cm.hot #颜色
             )

C = plt.contour(X,Y,f(X,Y)
             ,8 # 8+1条等高线
             ,colors = 'red'
             ,linewidth = 0.05
             )
             
plt.clabel(C
           ,inline = True #在线内加标注
           ,fontsize = 10
           )

plt.xticks(())
plt.yticks(())

plt.show()
```
